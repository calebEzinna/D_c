```python
import pandas as pd
```

```python
data = pd.read_csv('linkdin_Job_data.csv')
df = pd.DataFrame(data)
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df.shape)
```

```python
df.drop(['company_id', 'alumni', 'Hiring_person', 'linkedin_followers', 'Column1', 'hiring_person_link', 'job_details'], axis=1, inplace=True)
print(df)
```

```python
df.fillna({"no_of_employ" : df['no_of_employ'].mode()[0]}, inplace = True)
print(df.head)
```

```python
df['no_of_employees'] = df['no_of_employ'].str.split(' ').str[0]
print(df)
```

```python
df.drop(['no_of_employ'], axis=1, inplace = True)
print(df)
```

```python
df.fillna(method = 'ffill', inplace =True)
df.fillna(method = 'bfill', inplace =True)
print(df)
```

```python
df.isna().sum()
```

```python
print(df.head())
```

```python
pd.unique(df['full_time_remote'])
```

```python
df['full_time_remote'] = df['full_time_remote'].replace(['1-10 employees','11-50 employees'],['Executive','Director'])
df
```

```python
df['job'] = df['job'].str.split(',').str[0]
df
```

```python
df.isnull().sum()
```

```python
df.rename(columns = {'full_time_remote': 'employment_type'}, inplace = True)
#df
```

```python
x = {'Data Analyst - WFH':'Data Analyst', 'Software Engineer - Senior':'Software Engineer', '...':'Data Engineer'}
df['job'] = df['job'].replace(x)
```

```python
v = df[df['job'] == 'Data Analyst - WFH']
v
```

```python
h = df['job'].value_counts().head(20)
h
```

```python
df['no_of_employees'].value_counts()
```

```python
b = {'minutes':100, 'minute':101, 'hours':50, 'hour':69, 'seconds':100, 'days':57}
df['no_of_application'] = df['no_of_application'].replace(b)
```

```python
df['employment'] = df['employment_type'].str.split(' ').str.get(2)
df
```

```python
df.fillna({"employment" : df['employment'].mode()[0]}, inplace = True)
```

```python
df.drop(['employment_type'], axis=1, inplace = True)
```

```python
df.rename(columns = {'employment': 'employment_type'}, inplace = True)
```
